## OpenGL 

### OpenGL的特性

OpenGL 的 API 使用 C 语言编写的，因此 API 调用直接兼容 C 和 C++。 
对于其他十几种流行语言 OpenGL 也有着稳定的库，提供与 C 语言近似的性能。

### OpenGL & C++

使用 C++ 时，程序在 CPU 上运行，并包含 OpenGL 调用，我们将其称为 C++/OpenGL 应用程序。

C++/OpenGL 应用程序的一个重要任务是让程序员的 GLSL 代码运行与 GPU 上。

### GLSL

GLSL 是一种着色器语言。着色器语言主要运行于GPU上，在图形管线上下文中。

### 渲染上下文（Rendering Context）

渲染上下文主要包含了以下内容：

状态信息：颜色、材质、视图和模型矩阵、着色器程序、纹理映射等的信息。
命令队列：绘图命令和其他与图形处理相关的操作。

### OpenGL 管线

在管线中，将 3D 场景转换成 2D 画面的过程分割成了许多步骤。C++/OpenGL 应用程序发送图形数据到顶点着色器，随着管线处理，最终生成显示器上的像素点。

## 扩展库

实验中用到的扩展库，GLEW，GLM，SOIL2 和 GLFW。

### GLFW

OpenGL 并不是直接把图像绘制到计算机屏幕上，而是渲染到帧缓存区，然后由计算机把缓存区中的内容绘制到屏幕上的一个窗口中。
GLFW是目前最流行的支持这一部分工作的库之一

### GLEW（OpenGL extension wrangler）

扩展库是为了用来支持现代命令，可以执行复杂的代码。

### GLM（OpenGL Mathematics）

方便完成 3D 图形编程会大量使用的向量和矩阵代数运算。GLM 还包含了各种工具类，用来创建和使用和常见的 3D 图形结构，如透视和视角矩阵。

### SOIL2

SOIL 是一个纹理图像加载库，可以处理各种各样的图像文件格式，用来向图形场景中的对象添加“纹理”。
